```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
df=pd.read_csv("7. Udemy Courses.csv")
#print(df)
df.count()
df.isnull().sum()
df.head()
#IMP FOR STARTING A PROJECT ANALYSIS
```

```python
#Q. 1) What are all different subjects for which Udemy is offering courses ?
df.groupby('subject').count()
```

```python
df.groupby('subject').course_id.count().sort_values(ascending=False)
#df.groupby('Region').Confirmed.sum().sort_values(ascending=False).head(10)
```

```python
#show all the coures which r free f cost.
df[df.is_paid==False]
```

```python
#4) Show all the courses which are Paid.
df[df.is_paid==True]
```

```python
#Q. 5) Which are Top Selling Courses ?
df.sort_values('num_subscribers',ascending=False).head(5)
#Q. 6) Which are Least Selling Courses ?
df.sort_values('num_subscribers',ascending=True).head(5)
```

```python
#Q. 7) Show all courses of Graphic Design where the price is below 100 ?
df[(df['price']<'100') | (df['subject']=='Graphic Design')]
```

```python
#Q. 8) List out all the courses that are related to 'Python
df[df.course_title.str.contains('python')]
```

```python
#Q. 9) What are courses that were published in the year 2015 ?
df[df.published_timestamp=='2015']
```

```python
#Q. 10) What is the Max. Number of Subscribers for Each Level of courses ?
df.sort_values('num_subscribers',ascending=False)
```
